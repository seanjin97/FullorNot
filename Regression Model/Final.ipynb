# Function + Models

```python
#assuming time is the following format: Hour:Minute:Second

def xgboost_model(time,location):
    #preprocessing of the timing to get the bin
    time_list = time.split(':')
    bin_num = int(time[0])*6 + (int(time[1])//10+1)
    #retrieve the right dataframe
    df_name = loc_to_df[location]
    df_name = df_name + '.csv'
    df = pd.read_csv(df_name)
    
    ros = RandomOverSampler()
    x = np.array(df['bin']).reshape(-1,1)
    y = np.array(df['count']).reshape(-1,1)
    X,Y = ros.fit_sample(x,y)
    data_dmatrix = xgb.DMatrix(data=X,label=y)
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=45)
    xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.5, learning_rate = 0.25,
                max_depth = 20, alpha = 1, n_estimators = 25)
    xg_reg.fit(X_train,y_train)
    preds = xg_reg.predict(bin_num)
    return preds

print(xgboost_model('00:00:24','SIS L2 Along SRs'))
```
